```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv("coursera_courses.csv")
```

```python
df.head(10)
```

```python
df.drop(['Unnamed: 0'] , inplace = True , axis = 1 ) # here we drop the column Unnamed as it is useless 
```

```python
#second we want to change all letters to lower cases to help in analysis 
C = df.columns
for i in C :
    df[i] = df[i].apply(lambda text: " ".join(word.lower() for word in text.split()))
```

```python
# remove panctuations if exist in any column  
for i in C :
    if i != "course_rating":
        df[i] = df[i].str.replace('[^\w\s]','')
```

```python
print( len(df['_id'].unique()) ) # here _id is just used to label each course so it is  a redundant feature in our data set 
df.drop(['_id'] , inplace = True , axis = 1 )
```

```python
df.drop_duplicates() # remove the duplicates records if exist as it will be useless
```

```python
print( len ( df ) )
```

```python
for i in range(0 ,  len(df["course_type"])):
    df["course_type"][i] = df["course_type"][i].split(' ')[:-1]
```
