```python
from bs4 import BeautifulSoup

import requests

import pandas as pd 
```

```python
url = "https://boston.craigslist.org/search/sof"
```

```python
job_total = 0 

npo_job = {}

while True :
    
    response = requests.get(url)
    
    data_req = response.text 
    
    data_soup = BeautifulSoup( data_req , 'html.parser')
    
    jobs = data_soup.find_all( 'p' , {'class' : 'result-info'})
    
    for job in jobs :
        
        job_total += 1 
        
        title  = job.find( 'a' , {'class' : 'result-title'}).text
        
        location = job.find('span' , {'class' : 'result-hood'})
        
        location_tag = location.text if location else 'N/A'
        
        time = job.find( 'time' , {'class' : 'result-date'}).text
        
        link = job.find('a' , {'class' : 'result-title'} ).get('href')
        
        job_respons  = requests.get( link )
        
        job_data = job_respons.text
        
        job_soup = BeautifulSoup( job_data , 'html.parser')
        
        job_desc = job_soup.find('section' ,{'id' : 'postingbody'} ).text
        
        job_att_tag = job_soup.find('p' , {'class' : 'attrgroup'})
        
        job_att = job_att_tag.text if job_att_tag else 'N/A'
        
        print( "time is >> " , time ,'\n' ,"link is >> " ,  link , '\n' ,"location is >> ", location_tag ,'\n' , "title is >> ", title , '\n' , job_att )
        
        npo_job[job_total] = [title , location_tag  , time , link , job_att , job_desc ]
        
    url_tag = data_soup.find( 'a' , {'class': 'button next' } ) 
    
    if url_tag.get('href'):
        
        url = 'https://boston.craigslist.org' + url_tag.get('href')
        
    else:
        
        break
```

```python
jobs_data_frame = pd.DataFrame.from_dict(npo_job , orient = 'index' , columns = ['title' , 'location' , 'time' , 'link', 'job_att' , 'job_desc'] )
```

```python
jobs_data_frame.to_csv('jobs_data_frame.csv')
```

```python
# scraping APIS from https://www.programmableweb.com/category/all/apis
```

```python
url = 'https://www.programmableweb.com/category/all/apis' 

API_dic = {}

total_api = 0 

while True :
    
    data_req = requests.get(url)
    
    data_txt = data_req.text
    
    data_soup = BeautifulSoup(data_txt , 'html.parser')
    
    all_apis = data_soup.find_all('tr' , {'class' : 'odd views-row-first'})
    
    all_apis1 = data_soup.find_all('tr' , {'class' : 'even views-row-last'}) + all_apis 
    
    all_apis2 = data_soup.find_all('tr' , {'class' : 'even'}) + all_apis1
    
    all_apis3 = data_soup.find_all('tr' , {'class' : 'odd'}) + all_apis2
    
    #print(total_jobs)
    
    for api in all_apis3 :
        
        total_api += 1
        
        api_name = api.find('td' , {'class' : 'views-field views-field-pw-version-title'}).find('a').text #Api-Name 
        
        api_describtion_tag = api.find('td' , {'class' : 'views-field views-field-search-api-excerpt views-field-field-api-description hidden-xs visible-md visible-sm col-md-8'})
        
        api_describtion = api_describtion_tag.text if api_describtion_tag else 'N/A'
        
        api_category_tag = api.find('td' , {'class' : 'views-field views-field-field-article-primary-category'}).find('a')
        
        api_category = api_category_tag.text if api_category_tag else 'N/A'
        
        api_link = 'https://www.programmableweb.com/category/all/apis'+api.find('td' , {'class' : 'views-field views-field-pw-version-title'}).find('a').get('href') #Api-Name
        
        API_dic[total_api] = [ api_name , api_describtion , api_category , api_link ]
        
        
        #print("name :" , api_name , '\n' , "api desccription : " , api_describtion , '\n' , "api_category : " , api_category , '\n' , "api_link:" , api_link , '\n') 
        
    nxt_button = data_soup.find('a' , {'title':'Go to next page'}).get('href')
    
    #print(total_api)
    if nxt_button :
        
        url = 'https://www.programmableweb.com' + nxt_button 
        
    #print('https://www.programmableweb.com' + nxt_button)
    
    if total_api >= 500 : #take at most 500 APIS
        break
```

```python
# CHANGE TO DATA FRAME 
API_DATA_frame = pd.DataFrame.from_dict( API_dic , orient = 'index' , columns = ['API_NAME' , 'API_DESC' , 'API_CATEG' , 'API_LINK'] )
```
